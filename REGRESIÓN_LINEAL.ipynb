```
import numpy as np
import matplotlib.pyplot as plt
```

```
from sklearn.datasets import load_boston
```

```
boston = load_boston()
print(boston.DESCR)
```

Fórmula minimizar el error cuadrático medio (MCO): $\beta = (X^{T}X)^{-1}X^{T}Y$

```
x = np.array(boston.data[:,5])      # el 5 representa RM average number of rooms per dwelling del dataset
y = np.array(boston.target)         # esta función de numpy permite convertir los datos en un array

plt.scatter(x,y, alpha=0.3)         # alpha me deja ver con transparencia los puntos de la nube de puntos


#añadimos una columna de 1nos para el término independiente
x = np.array([np.ones(506), x]).T   # hacemos la matriz transpuesta para que de la forma (506, 2) 506 filas y dos columnas donde la primera esta llena de unos
B = np.linalg.inv(x.T @ x) @ x.T @ y # Fórmula del error cuadrático medio que está en el encabezado, el @ es para multiplicar matrices
plt.plot([4,9], [B[0]+B[1]*4, B[0]+B[1]*9], c="red") # imprime la línea de regresión lineal de acuerdo con el minimo error de cuadrados ordinarios
plt.show()
print("la forma de la matriz es: ", x.shape) 
print("el primer valor es el punto de intersección de la recta en cero y el segundo es el incremento entre los valores en el eje y: ")
print(B)
print("los datos de la matriz x son: ")
print(x)
```
